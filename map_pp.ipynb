```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Import State & District frm Geopandas
states=gpd.read_file(r'D:/Projects/phonepe_map/india_states.geojson')
districts=gpd.read_file(r'D:/Projects/phonepe_map/districts.geojson')
```

```python
# Plot the Map Sise by side
fig, (ax1, ax2)=plt.subplots(ncols = 2, figsize=(25,25))
states.plot(ax = ax1, cmap='YlGnBu',edgecolor="black",column='State_Name')
districts.plot(ax = ax2, cmap='RdYlGn',column='dtname')
ax1.axis('off')
ax2.axis('off')
```

```python
#Plot Multiple Layers
fig, ax=plt.subplots(figsize=(10,10))
state=states.to_crs(epsg = 7785)
district=districts.to_crs(epsg = 7785)
state.plot(ax = ax, cmap='prism',edgecolor='black',linewidth=1,column='State_Name')
district.plot(ax = ax, color='none', linewidth=0.2,edgecolor='black')
ax.axis('off')
```
